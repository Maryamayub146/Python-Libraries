```python
# <!-- lern subplot mean ek hi page py bar chart or line chart display krna  -->
# <!-- plt.subplot(nrows , ncolo , index) -->
import matplotlib.pyplot as plt
x = [1,2,3,4]
y = [10,14,15,17]

# 1 frist row , 2 mean 2 colm , 1st subplot
plt.subplot(1,2,1) 
plt.plot(x,y)
plt.title('line chart')

# 1 frist row , 2 colomn, 2nd subplot
plt.subplot(1,2,2)
plt.bar(x,y)
plt.title('bar chart')
plt.show()



```

```python
import pandas as pd
df = pd.read_csv('Downloads/sales_data_sample.csv' , encoding= 'latin1')

print(df.head())
print(df.tail())
```

```python
df.dropna(inplace=True)
# change data type
df['POSTALCODE'] = df['POSTALCODE'].astype(int)
df['PHONE'] = df['PHONE'].astype(int)
df['ORDERDATE'] = pd.i(df['ORDERDATE'], format='%d-%m-%Y')


print(df.info())
```

```python
# print(df.isnull().sum())
print(df.describe())
```

```python
print(df.isnull().sum())
```

```python
df.drop(['QUANTITYORDERED', 'QTR_ID','PRODUCTLINE'], axis=1, inplace=True)
```

```python
df['PHONE'] = df['PHONE'].str.replace(r'\D', '', regex=True)  # Removes all non-digit characters
df['PHONE'] = df['PHONE'].astype(int)

```

```python
print(df.info())
df['SALES'].value_counts()
```

```python
sns.set(style = 'whitegrid')
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,5))
df.groupby('SALES')[['QUANTITYORDERED','ORDERNUMBER']].sum().plot(ax=plt.gca())
plt.show()
```
